## print

### 基本参数




- *values: 要打印的对象
- sep: 分隔符，默认空格
- end: 结尾字符，默认换行
---------------------------------------------------

```python
print("aaa","vvv")
print("aaa""vvv")
print("aaa","vvv",sep="@",end=" ")
print("aaa","vvv",end="1")
>>aaa vvv
>>aaavvv
>>aaa@vvv aaa vvv1
```
---------------------------------------------------



- file: 输出流，默认 sys.stdout
----------
```python
print("hello", file=open("out.txt", "w"))
#不会在终端显示 "hello"，而是把 "hello" 写入到当前目录下的 out.txt 文件中。
"w" 是文件打开模式（mode）的一种，表示写入模式（write）。
     当你用 open("out.txt", "w") 打开文件时，如果文件不存在会自动创建，如果已存在会清空原内容再写入
"r"：只读模式（read），文件必须已存在
"a"：追加模式（append），写入内容会加到文件末尾
"b"：二进制模式（binary），如 "wb" 表示以二进制写入
```





### 格式输出

```python
name='wuyeye'
newname='xuanbao'
age=18
socre=90.5
# %s  字符串
# %d 整数 可补位
# %f 浮点数 默认6位 可补位
print("我的名字是%s和%s,年龄是%d,分数是%f" % (name,newname,age,socre) )
print("我的名字是%s和%s,年龄是%5d,分数是%.3f" % (name,newname,age,socre) )
print("我的名字是%s和%s,年龄是%03d,分数是%.1f" % (name,newname,age,socre) )


=======================================================
我的名字是wuyeye和xuanbao,年龄是18,分数是90.500000
我的名字是wuyeye和xuanbao,年龄是   18,分数是90.500
我的名字是wuyeye和xuanbao,年龄是018,分数是90.5
```
